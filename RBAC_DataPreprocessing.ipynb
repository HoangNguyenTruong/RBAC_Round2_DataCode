```python
merged_df_production_left = pd.merge(
    running_description_df,
    null_description_df,
    on=['PRODUCTIONDATE_DAY_LOC', 'SHIFT_NAME', 'LINE_NAME'],
    how='left',
    suffixes=('_running', '_null')
)

# Select the required columns
selected_columns = [
    ('PRODUCTIONDATE_DAY_LOC', 'PRODUCTIONDATE_DAY_LOC'),
    ('SHIFT_NAME', 'SHIFT_NAME'),
    ('LINE_NAME', 'LINE_NAME'),
    ('CO_TYPE_running', 'CO_TYPE'),
    ('SIZE_TYPE_running', 'SIZE_TYPE'),
    ('CREW_ID_running', 'CREW_ID'),
    ('AE_MODEL_CATEGORY_running', 'AE_MODEL_CATEGORY'),
    ('GOOD_PRODUCTION_QTY_null', 'GOOD_PRODUCTION_QTY'),
    ('REJECT_PRODUCTION_QTY_null', 'REJECT_PRODUCTION_QTY'),
    ('EFFECTIVE_RUNTIME_null', 'EFFECTIVE_RUNTIME'),
    ('PLANT_AVAILABLE_TIME_running', 'PLANT_AVAILABLE_TIME'),
    ('PRODUCTION_ORDER_RATE_running', 'PRODUCTION_ORDER_RATE')
]

# Create a new DataFrame with selected columns and renamed columns
result_df = merged_df_production_left[[col[0] for col in selected_columns]]
result_df.columns = [col[1] for col in selected_columns]

display(result_df.head())
```

```python
running_description_df = running_description_df.copy()
running_description_df['ID'] = range(1, len(running_description_df) + 1)
display(running_description_df.head())
```

```python
from datetime import timedelta

shift_order = ['Day-1', 'Afternoon-2', 'Night-3']
shift_mapping = {
    'Day-1': ('Afternoon-2', timedelta(days=0)),
    'Afternoon-2': ('Night-3', timedelta(days=0)),
    'Night-3': ('Day-1', timedelta(days=1))
}
```

```python
def get_next_shift_info(row):
    current_shift = row['SHIFT_NAME']
    # Convert to datetime object
    current_date = pd.to_datetime(row['PRODUCTIONDATE_DAY_LOC'])

    if current_shift in shift_mapping:
        next_shift, date_delta = shift_mapping[current_shift]
        next_date = current_date + date_delta
        return pd.Series([next_date, next_shift])
    else:
        return pd.Series([pd.NaT, None])

running_description_df[['JOIN_DATE', 'JOIN_SHIFT']] = running_description_df.apply(get_next_shift_info, axis=1)
display(running_description_df.head())
```

```python
running_description_df.shape
```

```python
# Convert the date column in null_description_df to datetime objects
null_description_df['PRODUCTIONDATE_DAY_LOC'] = pd.to_datetime(null_description_df['PRODUCTIONDATE_DAY_LOC'])

merged_production_logs = pd.merge(
    running_description_df,
    null_description_df[['PRODUCTIONDATE_DAY_LOC', 'SHIFT_NAME', 'LINE_NAME', 'ID']],
    left_on=['JOIN_DATE', 'JOIN_SHIFT', 'LINE_NAME'],
    right_on=['PRODUCTIONDATE_DAY_LOC', 'SHIFT_NAME', 'LINE_NAME'],
    how='inner',
    suffixes=('_running', '_null')
)

# Drop the redundant columns from the merged dataframe
merged_production_logs = merged_production_logs.drop(columns=['PRODUCTIONDATE_DAY_LOC_null', 'SHIFT_NAME_null'])

display(merged_production_logs.head())
```
